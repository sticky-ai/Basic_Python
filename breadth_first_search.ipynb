# Breadth First Search (깊이 우선 탐색)
* DFS와 함께 그래프 순회 알고리즘 중 하나

## 그래프 정의


```python
graph = {
    1: [2, 3, 4],
    2: [5],
    3: [5, 7],
    4: [],
    5: [6, 7],
    6: [],
    7: [3]
}

#   1
#  /|\
# 2 3 4
# |/\
# 5 7
# |
# 6


m = [
    [],
    [2, 3, 8],
    [1, 7],
    [1, 4, 5],
    [3, 5],
    [3, 4],
    [7],
    [2, 6, 8],
    [1, 7]
]
```

## Queue를 이용한 BFS

```python
from collections import deque

def bfs(graph, start_node, visited = []):
    queue = deque([start_node])
    visited.append(start_node)
    
    while queue:
        v = queue.popleft()
        print('Cur Node : {} | {}'.format(v, graph[v]))
        
        if graph[v]:
            for node in graph[v]:
                if node not in visited:
                    queue.append(node)
                    visited.append(node)
                    
            print('Queue    : {}'.format(queue))
            print('Visited  : {} \n'.format(visited))
            
        else:
            print('{} is end node \n'.format(v))
                
    return visited
```

```python
bfs(graph, 1)
```
